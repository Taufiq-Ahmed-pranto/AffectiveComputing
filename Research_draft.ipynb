#### Preprocess the data

```python
from bltk.langtools import Tokenizer, remove_stopwords
from bangla_stemmer.stemmer import stemmer

def preprocess_bangla_text(text):
    
    # Tokenize the text
    tokenizer = Tokenizer()
    words = tokenizer.word_tokenizer(text)

    # Remove stopwords
    filtered_words = remove_stopwords(words)
            
    # Stem the words
    stmr = stemmer.BanglaStemmer()
    stemmed_words = [stmr.stem(word) for word in filtered_words]
    
    stemmed_text = " ".join(stemmed_words)

    #Removes HTML syntaxes
    html_tag = re.compile(r'<.*?>')
    html_data = html_tag.sub(r'',stemmed_text)

    #Removes URL data
    url_clean = re.compile(r"https://\S+|www\.\S+")
    url_data = url_clean.sub(r'',html_data)

    #Removes Emojis
    emoji_clean= re.compile("["
                           u"\U0001F600-\U0001F64F"  # emoticons
                           u"\U0001F300-\U0001F5FF"  # symbols & pictographs
                           u"\U0001F680-\U0001F6FF"  # transport & map symbols
                           u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
                           u"\U00002702-\U000027B0"
                           u"\U000024C2-\U0001F251"
                           "]+", flags=re.UNICODE)
    preprocessed_text = emoji_clean.sub(r'',url_data)

    return preprocessed_text
```

```python
raw_train_data['cleaned_text'] = raw_train_data['text'].apply(preprocess_bangla_text)
```
